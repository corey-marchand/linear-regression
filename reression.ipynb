```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("./data.csv")
data = pd.read_csv("./data_two.csv")
print(f"Data Shape: {data.shape}")
print(data.head())
```

```python
print(data.columns)
```

```python
x = data['x'].values

y = data['y'].values

x_training_set, x_test_set, y_training_set, y_test_set = train_test_split(x,y)
```

```python
print(f"x_train shap {x_training_set.shape}")
print(f"y_train shape {y_training_set.shape}")
print(f"x_test shape {x_test_set.shape}")
print(f"y_test shape {y_test_set.shape}")
print(y_test_set)
```

```python
%matplotlib inline
plt.scatter(x_training_set, y_training_set,color='blue')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Training Linear regression')
plt.show()
```

```python
lm = LinearRegression()
lm.fit(x_training_set,y_training_set)
y_predict = lm.predict(x_test_set).reshape(-1,1)
print(f"Train accuracy {round(lm.score(x_training_set,y_training_set)*100,2)} %")
print(f"Train accuracy {round(lm.score(x_test_set,y_test_set)*100,2)} %")
```

```python
plt.scatter(x_training_set,y_training_set,color='red')
plt.plot(x_test_set,y_predict)
plt.xlabel("x label")
plt.ylabel("y lable")
plt.tiitle("Linear Regression Plotting")
plt.plot
```
